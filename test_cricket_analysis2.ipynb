```python
df_wickets = pd.read_csv("wickets.csv")
display(df_wickets)
```

```python
df_wickets = pd.read_csv('wickets.csv', encoding='unicode_escape', delimiter='\t')
display(df_wickets)
```

```python
display(df_wickets.head())
```

```python
print(df_wickets.columns)
```

### 1. Remove the column BBI, BBM

```python
df_wickets = df_wickets.drop(['BBI ','BBM '], axis=1)
display(df_wickets.head())
```

###### Removed two columns  'BBI' and 'BBM' from dataset

```python
print(df_wickets.columns)
```

##### Removed two columns 'BBI' and 'BBM' from dataset

### 2. How many players played for ICC?

```python
df_player = df_wickets['Player '].str.split("(", expand=True)

display(df_player.head(5))
```

#### Splitting the 'Player' column to get the information about 'Country'

```python
df_wickets = pd.concat([df_wickets, df_player], axis=1)

display(df_wickets.head())
```

##### Concatinating the new dataframe with the main dataframe

```python
df_wickets = df_wickets.drop('Player ', axis=1)

display(df_wickets.head())
```

###### Dropped the 'Player' column

```python
df_wickets = df_wickets.rename(columns={0: 'Player',
                                        1: 'Country'})

display(df_wickets.head())
```

##### Renamed the column names as player and country

```python
# remove the ")" from the 'Country' column
df_wickets['Country'] = df_wickets['Country'].str.replace(")","")

display(df_wickets.head())
```

```python
new_col_sequence = ['Player', 'Country','Span ', 'Mat ', 'Inns ', 'Balls ', 'Runs ', 'Wkts ', 'Ave ', 'Econ ',
       'SR ', '5 ', '10']
```

```python
df_wickets = df_wickets[new_col_sequence]

display(df_wickets.head())
```

##### The columns are rearranged 

```python
def icc_check(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
```

```python
df_wickets['played_for_ICC'] = df_wickets['Country'].apply(icc_check)

display(df_wickets.head(10))
```

##### Checked that a player had played for ICC or not

```python
df_wickets['played_for_ICC'].value_counts()
```

##### Showing how many players played for ICC

```python
df_wickets['Country'] = df_wickets['Country'].str.replace("ICC/", "")

display(df_wickets.head())
```

##### To get the country name, 'ICC/' is removed from the dataset 

### 3. How many different countries are present in this dataset? 

```python
df_wickets['Country'].value_counts()
```

##### There are 11 different type countries in the dataset

### 4. Which player(s) had played for the longest period of time?

```python
df_span = df_wickets['Span '].str.split("-", expand=True)
display(df_wickets.head(5))
```

##### splitted the 'Span' column based on the "-"

```python
df_wickets = pd.concat([df_wickets, df_span], axis=1)
display(df_wickets.head(5))
```

##### Concatinated the new dataframe with the main dataframe

```python
df_wickets = df_wickets.rename(columns={0: "start_year",
                                        1: "end_year"})
display(df_wickets.head(5))
```

##### Renamed the newly created column names to start_year and end_year

```python
df_wickets = df_wickets.drop("Span ", axis=1)

display(df_wickets.head())
```

##### Dropped the "Span" column because two new columns are created mentioning the information

```python
df_wickets.info()
```

##### Data type of start_year and end_year are object. We need those as integer

```python
df_wickets['start_year'] = df_wickets['start_year'].astype('int') 
df_wickets['end_year'] = df_wickets['end_year'].astype('int')

print(df_wickets.info())
```

##### Data type of start_year and end_year are integer which is required to get years_played

```python
df_wickets['years_played'] = df_wickets['end_year'] - df_wickets['start_year']

df_wickets= df_wickets.drop(['start_year', "end_year"], axis=1)

display(df_wickets.head(10))
```

##### Got the data mentioning years_played

```python
df_wickets.sort_values(by='years_played', ascending = False).head(5)
```

##### Imran Khan had played for the longest period of time i.e 21 years

### 5. Which player(s) had played for the shortest period of time?

```python
df_wickets.sort_values(by='years_played', ascending = False).tail(5)
```

##### GP Swann had played for the shortest period of time i.e 5 years
